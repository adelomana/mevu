```python
import pickle, os
```

# 0. user-defined variables

```python
jar_dir = 'results/'
```

# 1. read simulation jars

```python
jars = os.listdir(jar_dir)
total_number_of_jars = len(jars)
print(total_number_of_jars)
```

```python
%%time
gene_essentiality = {}

# iterate over samples
for jar in jars:
    #print(jar)
    
    jar_file = '{}{}'.format(jar_dir, jar)
    f = open(jar_file,'rb')
    [sampleID, result, single_ko_results] = pickle.load(f)
    f.close()
    
    print(sampleID, result)
    
    single_ko_results['ratio'] = single_ko_results['growth']/755.0032155506631
    essentials = single_ko_results[single_ko_results['ratio'] < 0.5]
    print(essentials.shape)
    
    essential_ids = []
    for element in essentials['ids']:
        essential_ids.append(list(element)[0])
    print(essential_ids)
    
    for element in essential_ids:
        if element not in gene_essentiality:
            gene_essentiality[element] = 1
        else:
            gene_essentiality[element] = gene_essentiality[element] + 1
            
print(gene_essentiality)
```

# 2. results

```python
print(gene_essentiality)
```

```python
for gene in gene_essentiality.keys():
    conditions_is_essential = gene_essentiality[gene]
    conditional_essentiality_frequency = conditions_is_essential / total_number_of_jars
    print(gene, conditions_is_essential, conditional_essentiality_frequency)
```
